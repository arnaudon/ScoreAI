```python
import numpy as np
a = np.random.uniform(0, 1, 10)
a
```

```python
print(f"Original array 'a':\n{a}\n")

# Compute the mean of 'a'
mean_a = np.mean(a)
print(f"Mean of 'a': {mean_a:.4f}\n")

# Compute the standard deviation of 'a'
std_a = np.std(a)
print(f"Standard deviation of 'a': {std_a:.4f}\n")

# Compute the sum of 'a'
sum_a = np.sum(a)
print(f"Sum of 'a': {sum_a:.4f}\n")

# Square each element in 'a'
squared_a = a**2
print(f"'a' squared element-wise:\n{squared_a}\n")

# Take the square root of each element in 'a'
sqrt_a = np.sqrt(a)
print(f"Square root of 'a' element-wise:\n{sqrt_a}\n")

# Multiply 'a' by a random scalar
random_scalar = np.random.uniform(0, 5)
scaled_a = a * random_scalar
print(f"'a' multiplied by {random_scalar:.4f}:\n{scaled_a}\n")
```
